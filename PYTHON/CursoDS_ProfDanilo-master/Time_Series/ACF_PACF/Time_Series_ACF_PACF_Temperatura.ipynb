# ACF e PACF

Determinando a autocorrelação e a autocorrelação parcial

Utilizaremos uma série temporal da temperatura na cidade australiana de Melbourne

* Importando biblitecas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

* Lendo CSV

```python
dados = pd.read_csv('daily-min-temperatures.csv')
```

* Exibindo cinco primeiras linhas

```python
dados.head()
```

* Definindo data como indice

```python
dados['Date'] = pd.to_datetime(dados['Date'])
dados = dados.set_index('Date')
```

* Graficando série temporal

```python
dados.plot()
plt.xticks(rotation=45)
plt.xlabel('Data')
plt.ylabel('Temperatura')
plt.tight_layout()
```

* Graficando autocorrelação

```python
from statsmodels.graphics.tsaplots import plot_acf
```

```python
plot_acf(dados.values,lags=80)
plt.xlabel('Lag')
plt.ylabel('ACF')
plt.show()
```

* Alterando range

```python
plot_acf(dados.values,lags=80)
plt.xlabel('Lag')
plt.ylabel('ACF')
plt.xlim([-1,10])
plt.show()
```

Encontramos que a lag do ACF tem valor 2

* Graficando autocorrelação parcial

```python
from statsmodels.graphics.tsaplots import plot_pacf
```

```python
plot_pacf(dados.values,lags=80)
plt.xlabel('Lag')
plt.ylabel('PACF')
plt.show()
```

* Alterando range

```python
plot_pacf(dados.values,lags=80)
plt.xlabel('Lag')
plt.ylabel('PACF')
plt.xlim([0,35])
plt.show()
```

Encontramos que a PACF tem lag igual a 2
